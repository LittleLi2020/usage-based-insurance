```python
import pandas as pd
import numpy as np
import time
```

```python
def timestamp2date(timestamp):
    timeArray = time.localtime(timestamp)
    formatted_time = time.strftime("%Y-%m-%d %H:%M:%S", timeArray)
    return formatted_time
```

```python
# read data
df = pd.read_csv(r"D:\Research\UBI车险\data\486_wuhan_test - LT.csv")
df[3716:3725]
```

```python
# convert timestamp
df["date"] = df["t"].apply(timestamp2date)
df = df[["t","date","VehSpdLgtA"]]
df[3716:3725]
```

```python
# calculate tangential acceleration
df["dVehSpdLgtA"] = df["VehSpdLgtA"].diff()
df["dt"] = df["t"].diff()
df["tg_acceleration"] = df["dVehSpdLgtA"] / df["dt"]
df[3716:3725]
```
